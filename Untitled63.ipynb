```python
# 1 What is the range of selling prices in the datasetD
# Calculate the minimum and maximum selling prices
selling_price_min = bike_data['selling_price'].min()
selling_price_max = bike_data['selling_price'].max()

# Calculate the range
selling_price_range = selling_price_max - selling_price_min

print(f"Minimum Selling Price: {selling_price_min}")
print(f"Maximum Selling Price: {selling_price_max}")
print(f"Range of Selling Prices: {selling_price_range}")
```

```python
# %B What is the median selling price for bikes in the datasetD
# Calculate the median selling price
median_selling_price = bike_data['selling_price'].median()

print(f"Median Selling Price: {median_selling_price}")
```

```python
# B What is the most common seller typeD
# Calculate the most common seller type
most_common_seller_type = bike_data['seller_type'].mode()[0]

print(f"Most Common Seller Type: {most_common_seller_type}")
```

```python
#  B How many bikes have driven more than 50,000 kilometersD
bikes_over_50k_km = bike_data[bike_data['km_driven'] > 50000].shape[0]

print(f"Number of bikes driven more than 50,000 kilometers: {bikes_over_50k_km}")
```

```python
# What is the average km_driven value for each ownership typeD
# Calculate the average km_driven for each ownership type
average_km_per_ownership = bike_data.groupby('owner')['km_driven'].mean()

print(average_km_per_ownership)
```

```python
# What proportion of bikes are from the year 2015 or olderD
# Filter bikes from the year 2015 or older
bikes_2015_or_older = bike_data[bike_data['year'] <= 2015]

# Calculate the proportion
proportion_2015_or_older = len(bikes_2015_or_older) / len(bike_data)

print(f"Proportion of bikes from 2015 or older: {proportion_2015_or_older:.2%}")
```

```python
#  What is the trend of missing values across the datasetD
# Count missing values in each column
missing_values = bike_data.isnull().sum()

# Total rows in the dataset
total_rows = len(bike_data)

# Calculate percentage of missing values for each column
missing_percentage = (missing_values / total_rows) * 100

# Combine results into a DataFrame for better readability
missing_summary = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage (%)': missing_percentage
})

print(missing_summary)
```

```python
# B What is the highest ex_showroom_price recorded, and for which bikeD
# Find the highest ex_showroom_price
highest_ex_showroom_price = bike_data['ex_showroom_price'].max()

# Filter the rows with the highest ex_showroom_price
bike_with_highest_price = bike_data[bike_data['ex_showroom_price'] == highest_ex_showroom_price]

print(f"Highest Ex-Showroom Price: ₹{highest_ex_showroom_price}")
print("Bike(s) with the highest ex-showroom price:")
print(bike_with_highest_price[['name', 'ex_showroom_price']])
```

```python
# 8;B What is the relationship between selling_price and km_driven for first-owner bikesD
import matplotlib.pyplot as plt

# Filter data for first-owner bikes
first_owner_bikes = bike_data[bike_data['owner'] == 'First Owner']

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(first_owner_bikes['km_driven'], first_owner_bikes['selling_price'], alpha=0.7)
plt.title('Relationship Between Selling Price and KM Driven (First-Owner Bikes)')
plt.xlabel('KM Driven')
plt.ylabel('Selling Price (₹)')
plt.grid(True)
plt.show()

# Calculate the correlation coefficient
correlation = first_owner_bikes['km_driven'].corr(first_owner_bikes['selling_price'])
print(f"Correlation Coefficient: {correlation}")
```

```python
# B Identify and remove outliers in the km_driven column using the IQR methodB
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = bike_data['km_driven'].quantile(0.25)
Q3 = bike_data['km_driven'].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Determine bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = bike_data[(bike_data['km_driven'] < lower_bound) | (bike_data['km_driven'] > upper_bound)]

# Remove outliers
bike_data_cleaned = bike_data[(bike_data['km_driven'] >= lower_bound) & (bike_data['km_driven'] <= upper_bound)]

print(f"Number of outliers: {len(outliers)}")
print(f"Dataset size after removing outliers: {len(bike_data_cleaned)}")
```

```python
#  Perform a bivariate analysis to visualize the relationship between year and selling_price
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = 'path_to_your_file.csv'  # Replace with the path to your CSV file
bike_data = pd.read_csv('BIKE DETAILS.csv')

# Inspect the data and clean if necessary
bike_data = bike_data[['year', 'selling_price']].dropna()  # Keep only relevant columns and drop missing values

# Ensure numeric data
bike_data['year'] = pd.to_numeric(bike_data['year'], errors='coerce')
bike_data['selling_price'] = pd.to_numeric(bike_data['selling_price'], errors='coerce')

# Drop rows with invalid numeric values
bike_data = bike_data.dropna()

# Calculate correlation
correlation = bike_data['year'].corr(bike_data['selling_price'])
print(f"Correlation between year and selling_price: {correlation:.2f}")

# Scatterplot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='year', y='selling_price', data=bike_data, scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
plt.title(f'Relationship between Year and Selling Price (Correlation: {correlation:.2f})')
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.grid(True)
plt.show()

# Boxplot for distribution by year
plt.figure(figsize=(12, 8))
sns.boxplot(x='year', y='selling_price', data=bike_data)
plt.title('Distribution of Selling Price by Year')
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# What is the average depreciation in selling price based on the bike's age (current year - manufacturing
# year)D
import pandas as pd
import datetime

# Load the CSV file (replace with your file path)
bike_data = pd.read_csv('BIKE DETAILS.csv')

# Add a column for the current year and calculate bike age
current_year = datetime.datetime.now().year
bike_data['bike_age'] = current_year - bike_data['year']

# Ensure numeric values for calculations
bike_data['selling_price'] = pd.to_numeric(bike_data['selling_price'], errors='coerce')
bike_data['bike_age'] = pd.to_numeric(bike_data['bike_age'], errors='coerce')

# Drop rows with invalid or missing values
bike_data = bike_data.dropna(subset=['bike_age', 'selling_price'])

# Group by bike age and calculate the average selling price
average_depreciation = bike_data.groupby('bike_age')['selling_price'].mean().reset_index()

# Rename columns for clarity
average_depreciation.columns = ['Bike Age', 'Average Selling Price']

# Print the result
print(average_depreciation)

# Optional: Visualize the depreciation trend
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(average_depreciation['Bike Age'], average_depreciation['Average Selling Price'], marker='o', linestyle='-')
plt.title('Average Depreciation in Selling Price by Bike Age')
plt.xlabel('Bike Age (Years)')
plt.ylabel('Average Selling Price')
plt.grid(True)
plt.show()
```

```python
# 8 B Which bike names are priced significantly above the average price for their manufacturing yearD
import pandas as pd

# Load the dataset (replace with your file path)
bike_data = pd.read_csv('BIKE DETAILS.csv')

# Ensure the required columns are present and clean the data
bike_data = bike_data[['name', 'year', 'selling_price']].dropna()

# Convert to numeric types
bike_data['year'] = pd.to_numeric(bike_data['year'], errors='coerce')
bike_data['selling_price'] = pd.to_numeric(bike_data['selling_price'], errors='coerce')

# Remove invalid rows
bike_data = bike_data.dropna()

# Calculate the average selling price for each manufacturing year
average_prices_by_year = bike_data.groupby('year')['selling_price'].mean().reset_index()
average_prices_by_year.columns = ['year', 'average_price']

# Merge the average prices back into the original dataset
bike_data = bike_data.merge(average_prices_by_year, on='year')

# Define a threshold for "significantly above" (e.g., 20% above the average price)
threshold_factor = 1.2
bike_data['is_above_average'] = bike_data['selling_price'] > (bike_data['average_price'] * threshold_factor)

# Filter for bikes significantly above the average price
above_average_bikes = bike_data[bike_data['is_above_average']]

# Display the results
print(above_average_bikes[['name', 'year', 'selling_price', 'average_price']])
```
